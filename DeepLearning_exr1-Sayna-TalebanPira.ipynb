```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Load the Fashion MNIST dataset
fashion_mnist = tf.keras.datasets.fashion_mnist

# Load the dataset into training and testing sets
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# Format of dataset
train_images.shape
```

```python
# Print number of classes
num_classes = len(set(train_labels))
print("Number of classes in Fashion MNIST dataset is:", num_classes)
```

```python
# Display 15 images in color mode
plt.figure(figsize=(15, 5))
for i in range(15):
    plt.subplot(3, 5, i + 1)
    plt.imshow(train_images[i])
    plt.colorbar()
    plt.axis('off')

plt.show()
```

```python
# Display 15 images in gray mode
plt.figure(figsize=(15, 5))
for i in range(15):
    plt.subplot(3, 5, i + 1)
    plt.imshow(train_images[i], cmap = 'gray')
    plt.colorbar()
    plt.axis('off')

plt.show()
```

```python
#Normalize dataset
train_images = train_images / 255
test_images = test_images / 255
```

```python
# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
model.fit(train_images, train_labels, epochs=10, verbose=2)
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_images, test_labels)
print(f"Test accuracy: {test_accuracy}")
```
